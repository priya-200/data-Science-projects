## Dataset:


> This is a dataset that is taken from Kaggle https://www.kaggle.com/datasets/mirichoi0218/insurance

## Problem statement:


> To find the charges of individual by using the other columns



## Content:
* age: age of primary beneficiary

* sex: insurance contractor gender, female, male

* bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,
objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

* children: Number of children covered by health insurance / Number of dependents

* smoker: Smoking

* region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

* charges: Individual medical costs billed by health insurance

## Import statements

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data science projects/medical expense /insurance.csv')
df.head()
```

```python
print("Number of data's available for us to use are :",len(df))
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df)
```

```python
# Features and labels

X = df.drop('charges',axis=1)
y = df.charges
```

```python
y
```

```python
# spliting the train and test set

np.random.seed(42)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Building a neural network

tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(50),
    #tf.keras.layers.Dense(5),
    tf.keras.layers.Dense(1)
])

model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ['mae']
              )

model.fit(X_train,y_train,epochs = 100)
```

```python
# Check result of model in test data

model.evaluate(X_test,y_test)
```

```python
# Create a better model

tf.random.set_seed(24)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_1.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(lr=0.001),
    metrics = ['mae']
)

model_1.fit(X_train,y_train,epochs=100)
```

```python
model_1.evaluate(X_test,y_test)
```

```python
# Model_2

model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_2.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(lr=0.001),
    metrics = ['mae']
)

model_2.fit(X_train,y_train,epochs=300)
```

```python
model_2.evaluate(X_test,y_test)
```

## Preprocessing data(Normalization,Standardization)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data science projects/medical expense /insurance.csv')
df.head()
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split

ct = make_column_transformer(
    (MinMaxScaler(),['age','bmi','children']),
    (OneHotEncoder(handle_unknown="ignore"),['sex','region','smoker'])
)

# Create X and y
X = df.drop('charges',axis=1)
y= df.charges

# Train test split
np.random.seed(42)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

# Fit the column transformer to our training data
ct.fit(X_train)

# Transform training and test data to its normalised and onehot encoded form

X_train_n = ct.transform(X_train)
X_test_n = ct.transform(X_test)
```

```python
# What does our data look like

X_test_n[0]
```

```python
# What is the difference in shape of the data

X_train.shape,X_train_n.shape
```

```python
# Build a neural network

tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_3.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(lr=0.001),
    metrics = ['mae']
)

model_3.fit(X_train_n,y_train,epochs=300)
```

```python
# Evaluate

model_3.evaluate(X_test_n,y_test)
```

```python
# Build a neural network

tf.random.set_seed(42)

model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_4.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(lr=0.01),
    metrics = ['mae']
)

model_4.fit(X_train_n,y_train,epochs=300)
```

```python
model_4.evaluate(X_test_n,y_test)
```
